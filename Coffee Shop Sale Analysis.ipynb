```python
ax = sns.countplot(x= 'product_category', data =df)
for bars in ax.containers:
    ax.bar_label(bars)
    sns.set(rc= {'figure.figsize': (20,5)})
```

```python
as you can see the in the plot coffee buyers are more than other catagories 
```

```python
df.groupby(['product_category'], as_index=False)['Sales'].sum().sort_values(by= 'Sales', ascending= False)
```

```python
df.groupby(['product_category'], as_index=False)['Sales'].sum().sort_values(by= 'Sales', ascending=False).head(10)
sns.barplot(x= 'Sales', y= 'product_category', data= df)
plt.figure(figsize=(30, 15), dpi=100)
```

```python
df.groupby(['store_location'], as_index=False)['Sales'].sum().sort_values(by= 'Sales', ascending=False).head(10)
sns.barplot(x= 'store_location', y= 'Sales', data= df)
sns.set(rc={'figure.figsize': (10, 5)})
```

```python
df.groupby(['product_type'], as_index=False)['Sales'].sum().sort_values(by='Sales', ascending=False).head(5)
sns.barplot(x= 'product_type', y= 'Sales', data=df)
sns.set(rc={'figure.figsize': (40, 5)})
```

```python
df.groupby(['product_category'], as_index=False)['transaction_qty'].sum().sort_values(by='transaction_qty', ascending=False).head(5)
sns.barplot(x= 'product_category', y= 'transaction_qty', data=df)
sns.set(rc={'figure.figsize': (15, 5)})
```
