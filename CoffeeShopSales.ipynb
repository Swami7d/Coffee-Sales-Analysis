```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('coffee shop sales dataset.csv', encoding = 'unicode_escape')
# To Avoide errors in encoding use unicode_escape
```

```python
ax = sns.countplot(x= 'store_location', data= df)
for bars in ax.containers:
    ax.bar_label(bars)
    sns.set(rc= {'figure.figsize': (15,10)})
```

```python
# This Chart shows no.of orders in each store location, where we can see the "Hell's Kitchen" has more orders than the other two locations
```

```python

ax = sns.countplot(x= 'product_category', data =df)
for bars in ax.containers:
    ax.bar_label(bars)
    sns.set(rc= {'figure.figsize': (20,5)})
```

```python
# from this chart you can see that coffee catagory has more demand than other catagories 
```

```python
product_type_counts = df['Month Name'].value_counts()
plt.pie(product_type_counts, labels=product_type_counts.index, autopct='%1.1f%%', startangle=200)
```

```python
# This shows the percentage of distribution of sales in month
```

```python
df.groupby(['Sales'], as_index=False)['Month'].sum().sort_values(by= 'Month', ascending=False).head(5)
sns.lineplot(x= 'Month', y= 'Sales', data= df)
sns.set(rc={'figure.figsize': (10, 5)})
plt.figure(figsize=(30, 15), dpi=200)
```

```python
# This shows Sales by month where we can see that 6th months its increased
```

```python
df.groupby(['product_category'], as_index=False)['Total_Bill'].sum().sort_values(by= 'Total_Bill', ascending=False).head(5)
sns.barplot(x= 'product_category', y= 'Total_Bill', data= df)
sns.set(rc={'figure.figsize': (20, 10)})
plt.figure(figsize=(30, 15), dpi=200)
for bars in ax.containers:
    ax.bar_label(bars)
```
